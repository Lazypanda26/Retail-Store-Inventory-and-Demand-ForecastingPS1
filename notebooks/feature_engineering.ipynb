```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("../data/sales_data.xls")
```

```python
df = df.sort_values(
    ["Store ID", "Product ID", "Date"]
)
df.head()
```

Demand Lags

```python
group_cols = ["Store ID", "Product ID"]

df["Demand_Lag_1"] = (
    df.groupby(group_cols)["Demand"]
    .shift(1)
)

df["Demand_Lag_7"] = (
    df.groupby(group_cols)["Demand"]
    .shift(7)
)

df["Demand_Lag_30"] = (
    df.groupby(group_cols)["Demand"]
    .shift(30)
)

df.head()
```

Price lag

```python
df["Price_Lag_1"] = (
    df.groupby(group_cols)["Price"]
    .shift(1)
)
df.head()
```

Rolling demand mean

```python
df["Demand_Mean_7"] = (
    df.groupby(group_cols)["Demand"]
    .transform(
        lambda x: x.rolling(7).mean()
    )
)

df["Demand_Mean_30"] = (
    df.groupby(group_cols)["Demand"]
    .transform(
        lambda x: x.rolling(30).mean()
    )
)

df.head()
```

Rolling demand volatility

```python
df["Demand_Std_7"] = (
    df.groupby(group_cols)["Demand"]
    .transform(
        lambda x: x.rolling(7).std()
    )
)

df["Demand_Std_30"] = (
    df.groupby(group_cols)["Demand"]
    .transform(
        lambda x: x.rolling(30).std()
    )
)

df.head()
```

Rolling max

```python
df["Demand_Max_7"] = (
    df.groupby(group_cols)["Demand"]
    .transform(
        lambda x: x.rolling(7).max()
    )
)
df.head()
```

All other useful statistics

```python
df["Revenue"] = df["Units Sold"] * df["Price"] * (1 - (df["Discount"] / 100))
df["Inventory Value"] = df["Inventory Level"] * df["Price"]
df["Lost Demand"] = np.maximum(df["Demand"] - df["Units Sold"], 0)
df["Stockout"] = df["Inventory Level"] < df["Demand"]
df["Overstock"] = df["Inventory Level"] > (2 * df["Demand"])
df["Sell Through Rate"] = df["Units Sold"] / np.maximum(df["Inventory Level"], 1)
df["Inventory Turnover"] = df["Units Sold"] / np.maximum(df["Inventory Level"], 1)
df["Coverage Days"] = df["Inventory Level"] / np.maximum(df["Demand"], 1)
df["Price Gap"] = df["Price"] - df["Competitor Pricing"]
df["Gross Profit (Proxy)"] = df["Revenue"] - (df["Units Sold"] * df["Price"] * 0.4)

df.head()
```

RAG Embeddings

```python
df["rag_text"] = (
    "Date: " + df["Date"].astype(str)
    + ". Store: " + df["Store ID"].astype(str)
    + ". Product: " + df["Product ID"].astype(str)
    + ". Category: " + df["Category"].astype(str)
    + ". Region: " + df["Region"].astype(str)
    + ". Demand: " + df["Demand"].astype(str)
    + ". Units Sold: " + df["Units Sold"].astype(str)
    + ". Inventory: " + df["Inventory Level"].astype(str)
    + ". Price: " + df["Price"].astype(str)
    + ". Discount: " + df["Discount"].astype(str)
    + ". Promotion: " + df["Promotion"].astype(str)
    + ". Weather: " + df["Weather Condition"].astype(str)
    + ". Seasonality: " + df["Seasonality"].astype(str)
    + ". Epidemic: " + df["Epidemic"].astype(str)
)
df.head()
```

Handling missing values

```python
df = df.fillna(0)
```

test and train split

```python
df["Date"] = pd.to_datetime(df["Date"])

df = df.sort_values("Date").reset_index(drop=True)

split_idx = int(len(df) * 0.8)

train_df = df.iloc[:split_idx]
test_df = df.iloc[split_idx:]

print(train_df["Date"].min())
print(train_df["Date"].max())

print(test_df["Date"].min())
print(test_df["Date"].max())
```

Making train and test csv files

```python
train_df.to_csv("train_data.csv", index=False)
test_df.to_csv("test_data.csv", index=False)
```

Saving fully feature engineered dataframe

```python
df.to_csv("processed_sales_data.csv", index=False)
```
